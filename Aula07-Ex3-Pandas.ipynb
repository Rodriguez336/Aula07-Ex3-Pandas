c) Visualize o cabeçalho dos dados;

d) Visualize as últimas linhas;

e) Escreva um programa para renomear o cabeçalho da tabela com os seguintes nomes:




```python
# Importando a biblioteca do Pandas

import pandas as pd

# Mapear o caminho e o nome do documento

caminho = './arquivos/Vendas.xlsx'

# Atribuir os dados a um dataframe
df = pd.read_excel(caminho)

# Visualizar o cabeçalho
df.head()

```

```python
# Renomear as colunas

colunas = ['data','vendedor','cargo','cliente','uf','vendas']

df.columns = colunas

df.head()
```

```python
# Selecionar a coluna de interesse

df_cargos = df['cargo'].drop_duplicates()

cargos = df_cargos.to_list()
cargos

```

```python
for cargo in cargos:
    df_final = df[df['cargo'] == cargo]
    caminho = './relatorios-exercicio-03/'
    arquivo = cargo + '.xlsx'
    df_final.to_excel(caminho + arquivo)
   
```

```python
df_clientes = df['cliente'].drop_duplicates()

clientes = df_clientes.to_list()

for cliente in clientes:
    df_final = df[df['cliente'] == cliente]
    caminho = './relatorios-exercicio-03/'
    arquivo = cliente + '.xlsx'
    df_final.to_excel(caminho + arquivo)

```
